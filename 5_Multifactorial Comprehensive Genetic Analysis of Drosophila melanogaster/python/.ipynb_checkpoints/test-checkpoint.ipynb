```python
import os
import pandas as pd
import numpy as np
import flydataclass as fc
from itertools import product
from scipy.stats import chi2_contingency

f1_dir = "./data/f1"
f2_dir = "./data/f2"
```

# Load data
* F1

```python
csv_files = [
    file for file in os.listdir(f1_dir) if file.endswith('.csv')
]

f1_data = []
for file in csv_files:
    if file == "data_全体汇总.csv":
        continue
    file_path = os.path.join(f1_dir, file)
    df = pd.read_csv(file_path)
    numpy_df = df.to_numpy()
    for row in numpy_df[2:, ]:
        f1_data.append(fc.GroupData(row, 1))
```

* F2

```python
csv_files = []
for item in os.listdir(f2_dir):
    item_path = os.path.join(f2_dir, item)
    if os.path.isdir(item_path):
        for file in os.listdir(item_path):
            if file.endswith(".csv"):
                filr_path = os.path.join(item_path, file)
                csv_files.append(filr_path)

f2_data = []
for file in csv_files:
    if file == "./data/f2\\all data\\合计.csv":
        continue
    df = pd.read_csv(file)
    df = df.iloc[2:10, 2:6].join(df.iloc[2:10, 9:13])
    numpy_df = df.to_numpy()
    f2_data.append(fc.GroupData(numpy_df, 2))
```

## filter data

```python
f1_orth_valid = 0
f1_comp_valid = 0
f2_valid = 0

for group_data in f1_data:
    if group_data.filter_data("orth"):
        f1_orth_valid = f1_orth_valid + 1
    if group_data.filter_data("comp"):
        f1_comp_valid = f1_comp_valid + 1
print("f1_orth_valid:", f1_orth_valid)
print("f1_comp_valid:", f1_comp_valid)

for group_data in f2_data:
    if group_data.filter_data():
        f2_valid = f2_valid + 1
print("f2_both_valid:", f2_valid)
```

```python
import matplotlib.pyplot as plt

def plot_hist(data, trait_list, phenotype, type):
    f2_orth_grey_red = []
    for item in data:
        all = item.get_count(type)
        grey_red = item.get_motify_specific_count(trait_list, phenotype, type)
        if all == 0 or all is None:
            continue
        else:
            f2_orth_grey_red.append(grey_red/all)
    f2_orth_grey_red = np.array(f2_orth_grey_red)
    plt.hist(f2_orth_grey_red, bins=10)
    plt.show()

from sklearn.cluster import KMeans
# 设置环境变量 OMP_NUM_THREADS 为 1
os.environ["OMP_NUM_THREADS"] = "1"
#x = np.random.random(10000).reshape(-1,1)
#km = KMeans(n_clusters=5)
#km.fit(x)
#print(km.cluster_centers_)

def k_means(data, trait_list, phenotype, type):
    target_data = []

    for item in data:
        all_count = item.get_count(type)
        target_count = item.get_motify_specific_count(trait_list, phenotype, type)

        if all_count == 0 or all_count is None:
            continue
        else:
            target_data.append(target_count / all_count)

    if not target_data:
        print("No valid data for clustering.")
        return

    target_data = np.array(target_data)

    # 初始化KMeans，n_clusters为聚类数目，n_init为初始化中心点的次数
    km = KMeans(n_clusters=3, n_init=10)
    km.fit(target_data.reshape(-1, 1))

    # 找到具有最多数据点的簇的索引
    largest_cluster_index = np.argmax(np.bincount(km.labels_))

    # 提取最大簇的数据
    largest_cluster_data = target_data[km.labels_ == largest_cluster_index]

    # 打印最大簇的中心点
    print("Centroid of the largest cluster:", km.cluster_centers_[largest_cluster_index])

    # 打印最大簇的数据
    print("Data points in the largest cluster:", largest_cluster_data)

# 示例调用
# k_means(data, trait_list, phenotype, type)

plot_hist(f2_data, [1,0,0,0,1],[1,1], "comp")
k_means(f2_data, [1,0,0,0,1],[1,1], "comp")
plot_hist(f2_data, [1,0,0,0,1],[0,1], "comp")
k_means(f2_data, [1,0,0,0,1],[0,1], "comp")
plot_hist(f2_data, [1,0,0,0,1],[1,0], "comp")
k_means(f2_data, [1,0,0,0,1],[1,0], "comp")
plot_hist(f2_data, [1,0,0,0,1],[0,0], "comp")
k_means(f2_data, [1,0,0,0,1],[0,0], "comp")
```

```python
f2_df = pd.DataFrame()
bin_list = list(product([0,1], repeat=5))
my_list = []

for item in f2_data:
    temp = []
    for bin in bin_list:
        num = item.get_specific_count(bin, "comp")
        all = item.get_count("comp")
        if all == 0 or all is None:
            temp.append(0)
        else:
            temp.append(num/all)
    my_list.append(temp)

f2_df = pd.DataFrame(my_list, columns=bin_list)

print(f2_df)

## Initialize the KMeans model with the desired number of clusters
#num_clusters = 3  # You can adjust this based on your requirements
#kmeans_model = KMeans(n_clusters=num_clusters)
#
## Fit the model to your data
#kmeans_model.fit(f2_df)

# Try different values of k
k_values = range(1, 11)
inertias = []

for k in k_values:
    kmeans_model = KMeans(n_clusters=k)
    kmeans_model.fit(f2_df)
    inertias.append(kmeans_model.inertia_)

# Plot the Elbow graph
plt.plot(k_values, inertias, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()

kmeans_model = KMeans(n_clusters=2)
kmeans_model.fit(f2_df)

# Get the labels assigned to each data point
labels = kmeans_model.labels_

# Add the labels to your DataFrame
f2_df['Cluster'] = labels

# Now f2_df contains an additional column 'Cluster' indicating the cluster assignment for each data point
print("K-Means Results:")
print(f2_df[['Cluster']])

# Get the centroids of each cluster
centroids = kmeans_model.cluster_centers_

# Print the centroids
print("Cluster Centroids:")
print(centroids)

```
